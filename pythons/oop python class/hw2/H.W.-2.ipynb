```python
import numpy as np
import pandas as pd
```

```python
# Read the data file using pandas.read_csv
# Store the dataframe in a object
df = pd.read_csv(r'seattle-weather.csv', encoding='utf8', engine='python')
```

```python
# Display the data
df
```

```python
# Extracting the year from date column using pandas function DatetimeIndex
df['year'] = pd.DatetimeIndex(df['date']).year
```

```python
# Displaying the value count from year column to confirm the values make sense
df['year'].value_counts()
```

- Use mask operation to display the data of 2013, 2014, and 2015 respectively

```python
# Accessing 2013 year dataframe by using comparison operator of equal to
# Putting the condition data in df dataframe using index
# Above two steps are called masking
```

```python
df[df['year'] == 2013]
```

```python
df[df['year'] == 2014]
```

```python
df[df['year'] == 2015]
```

```python
df
```

- How many number of days without rain in 2013, and 2014?

```python
print(f'No rainy days in 2013: {np.sum(df[df["year"] == 2013]["precipitation"] == 0)}')
print(f'No rainy days in 2014: {np.sum(df[df["year"] == 2014]["precipitation"] == 0)}')
```

- The other way to calculate no rainy days in 2013 and 2014

```python
df1 = df[((2013 == df["year"]) | (df["year"] == 2014))]
```

```python
print(f'No rainy days in 2013 and 2014: {np.sum(df1["precipitation"] == 0)}')
```

- Calculate number of days with rain in 2014, and 2015?

```python
print(f'Number of rainy days in 2014: {np.sum(df[df["year"] == 2014]["precipitation"] != 0)}')
print(f'Number of rainy days in 2015: {np.sum(df[df["year"] == 2015]["precipitation"] != 0)}')
```

- The other way to calculate number of rainy days in 2014 and 2015

```python
df2 = df[((2014 == df["year"]) | (df["year"] == 2015))]
```

```python
print(f'Number of rainy days in 2014 and 2015: {np.sum(df2["precipitation"] != 0)}')
```

- How many rainy days were there with more than 5mm of rain in 2013, and 2015?

```python
print(f'Number of rainy days with more than 5mm of rain in 2013: {np.sum(df[df["year"] == 2013]["precipitation"] > 5)}')
print(f'Number of rainy days with more than 5mm of rain in 2015: {np.sum(df[df["year"] == 2015]["precipitation"] > 5)}')
```

- The other way to calculate rainy days with more than 5mm of rain in 2013, and 2015?

```python
df3 = df[((2013 == df["year"]) | (df["year"] == 2015))]
```

```python
print(f'Number of rainy days with more than 5mm of rain in 2013 and 2015: {np.sum(df3["precipitation"] > 5)}')
```

- How many rainy days were there with less than 15mm of rain in 2013, 2014, and 2015 respectively?

```python

print(f'Number of rainy days with less than 15mm of rain in 2013: '
      f'{np.sum((df[df["year"] == 2013]["precipitation"] < 15) & (df[df["year"] == 2013]["precipitation"] > 0))}')
print(f'Number of rainy days with less than 15mm of rain in 2014: '
      f'{np.sum((df[df["year"] == 2014]["precipitation"] < 15) & (df[df["year"] == 2014]["precipitation"] > 0))}')
print(f'Number of rainy days with less than 15mm of rain in 2015: '
      f'{np.sum((df[df["year"] == 2015]["precipitation"] < 15) & (df[df["year"] == 2015]["precipitation"] > 0))}')
```
