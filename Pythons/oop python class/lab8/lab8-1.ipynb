### Figures

- Plots in matplotlib resides in a figures object
- We can create a new figure with `plt.figure()`

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
plt.plot(np.random.randn(50).cumsum(), 'g--')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

ax1.hist(np.random.randn(100), bins=40, color='m', alpha=0.33)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
plt.plot(np.random.randn(50).cumsum(), 'g--')
```

```python
# Adjust the space around sub_plots
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(300), bins=50, color='g', alpha=0.5)
        
plt.subplots_adjust(wspace=1, hspace=0.5)   
```

### Color, marker, and line styles

```python
# Way of using the color and style
fig = plt.figure()
plt.plot(np.random.randn(30).cumsum(), linestyle='--', color='g')
```

### Line plot

- Using Series
- Using DataFrame

#### Series line plot

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

#### DataFrame line plot

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0), columns=['A', 'B', 'C', 'D'], index=np.arange(0, 100, 10))
df.plot()
```

### Bar plot

- Series bar plot
- DataFrame bar plot
- Stacked bar plot horizontal / vertical

```python
# Series bar plot
fig, axes = plt.subplots(2, 1, figsize =(12, 10))
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0], color='g', alpha=0.7)
data.plot.barh(ax=axes[1], color='g', alpha=0.7)
```
